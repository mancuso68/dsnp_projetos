##**Um pouco da História**
Descoberto em janeiro de 1502 por navegadores portugueses que, ao avistarem a entrada da *Baia de Guanabara*, acreditavam ter descoberto a foz de um grande rio. Deram ao lugar o nome de *Rio de Janeiro*.

Os índios que habitavam a região eram da etnia dos *Tamoios* e chamavam o local de *Guanabara e Niteroi*.

* Guanabara - significava "*seio*" ou "*nascedouro*" do mar, chamavam assim o fundo da baia onde desaguavam a maioria dos rios dado aos nativos a impressão de que o mar nascia ali.
* Niteroi - significava *"água escondida"* ou *"mar morto"*, por apresentar águas calmas na parte onde hoje se encontra a cidade com mesmo nome.

A região ribeirinha da atual cidade do Rio de Janeiro onde os portugueses ergueram uma casa (próximo a praia hoje chamada de Flamengo) era chamada pelos nativos como *Carioca* - branco(*cari*) e casa(*oca*) - Casa de Branco.

O Rio de Janeiro tem muita história para ser contada ontes de se tornar uma cidade com altos índices de violência.

No Livro **O Rio Antigo** de Eduardo Camões (hoje encontrado apenas em sebos), o autor apresenta diversas pinturas feitas por ele que são norteadas pela história dessa cidade maravilhosa.

## **Informações Estatísticas da Violência no Rio de Janeiro**

Antes de falar sobre a análise estatística dos dados, cabe explicar algumas definições:
* Furto - Subtração de um bem alheio sem contato ou violência com a vítima.
* Roubo - Subtração de um bem alheio mediante ameaça ou violência.
* Homicídio culposo - Praticado sem a intenção de matar.
* Homicídio doloso - Com dolo, praticado com a intenção de matar.

A análise levará em conta as colunas de roubo e furto de veículos, roubo em coletivos, veículos recuperados e homicídos dolosos.

Abaixo um resumo estatístico de cada uma das colunas do DataFrame.

```
#Resumo estatístico do DataFrame
df.describe()
```

As próximas linhas irão apresentar os dados estatísticos de algumas colunas. As informações calculadas aqui abaixo podem ser visualizadas no resumo acima. O objetivo maior aqui é mostrar como realizar esses cálculos em *python*, usando a biblioteca *Pandas*.

* Médias, Medianas e Desvio Padrão das seguintes variáveis (colunas):
  * `roubo_veiculo`
  * `furto_veiculos`
  * `recuperacao_veiculos`
* Percentual de carros recuperados em relação aos carros roubados + carros furtados:
  * $\frac{\text{recuperacao_veiculos}}{\text{roubo_veiculo} + \text{furto_veiculos}}$

* Maior e menor valor de Homicídios dolosos


```
#Calculando as médias e medianas de Roubo e furto de veículos
med_roubo = df.roubo_veiculo.mean()
med_furto = df.furto_veiculos.mean()
men_roubo = df.roubo_veiculo.median()
men_furto = df.furto_veiculos.median()

#Calculando a média e mediana de recuperação de veículos
med_recup = df.recuperacao_veiculos.mean()
men_recup = df.recuperacao_veiculos.median()

print('Média de roubo de veículos:      {:.2f}'.format(med_roubo))
print('Mediana de roubo de veículos:    {:.2f}'.format(men_roubo))
print('Desvio padrão                    {:.2f}'.format(df.roubo_veiculo.std()))
print("----")
print("Média de furto de veículos:      {:.2f}".format(med_furto))
print("Mediana de furto de veículos:    {:.2f}".format(men_furto))
print('Desvio padrão                    {:.2f}'.format(df.furto_veiculos.std()))
print("----")
print("Média de recuperação de veículos   {:.2f}".format(med_recup))
print("Mediana de recuperação de veículos {:.2f}".format(men_recup))
print('Desvio padrão                      {:.2f}'.format(df.recuperacao_veiculos.std()))
print("----")
#Calculando o percentual de veículos rrecuperados em relação ao total de veículos roubados e furtados
percent01 = (med_recup/(med_roubo + med_furto))*100
print('Percentual de carro recuperados em relação aos roubados e furtados: {:.2f}%'.format(percent01))
print("----")
#Imprimindo o maior e menor valor entre os dados de Homicídio doloso

print("Maior valor de homicídios dolosos: \t{}".format(df.hom_doloso.max()))
print("Menor valor de homicídios dolosos: \t{}".format(df.hom_doloso.min()))
print("Ano e mês do Maior valor de Homicícios dolosos ",df[['vano','mes']][df['hom_doloso']==df['hom_doloso'].max()])
print("Ano e mês do Menor valor de Homicícios dolosos ",df[['vano','mes']][df['hom_doloso']==df['hom_doloso'].min()])
```

###**Alterando o Índice do Dataframe**
Os dados estão organizados no DataFrame condorme os campos *vano* e *mes*. Como pode ser observado, trata-se de uma série temporal, porém não existe um campo único indicando da data.

Com isso, será criada uma coluna de data contendo o Ano, Mês e para o dia será considerado o primeiro dia do mês. Depois essa coluna será transformada em índice do DataFrame.

###**Média Móvel Simples**

A média móvel simples serve para suavizar os dados de cada coluna para formar um indicador de tendência. Serão considerados 5 valores para cálculo, ou seja, a cada período é calculada a média entre o valor do período mais os valores dos 4 últimos periodos.

```
#criando os campos de média movel para as 4 colunas
df['mms_roubo_veiculo']=0.0
df['mms_furto_veiculos']=0.0
df['mms_roubo_em_coletivo']=0.0
df['mms_roubo_transeunte']=0.0

#função para preenchimento dos campos
  #Coluna -> nome da coluna que vai calcular a média Móvel Simples
  #Indice -> quantidade de períodos a ser considerado no cálculo da Média Móvel Simples
def mms(coluna,indice):
  colmms='mms_'+coluna
  indice=indice-1

  for i in range(indice,df.shape[0]):   #percorrendo o Dataframe
    mms=0

    for n in range(i-indice,i):   #percorrendo as entradas para cálculo da Média Movel Simples
      mms=mms+df[coluna][n]

    mms=mms+df[coluna][i]   #somando a entrada atual
    mms=mms/5               #Calculando a Média Móvel Simples   
    df[colmms][i]=mms       #Atribuindo o valor da MMS na coluna

#completando as colunas com as Médias Móveis Simples 
mms('roubo_veiculo',5)
mms('furto_veiculos',5)
mms('roubo_em_coletivo',5)
mms('roubo_transeunte',5)
```

Após calcular as médias simples de cada coluna e inserir os valores no Dataframe, serão plotados novamente cada gráfico, incluindo a MMS.

```
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1,figsize=(10,13))

#Roubo de Veículos
df['roubo_veiculo'].plot(kind='line', color='gray', linewidth=3, ax=ax1)
df['mms_roubo_veiculo'].plot(kind='line',color='blue', linewidth=2, ax=ax1)

ax1.set_title('Roubos de Veículos')
ax1.set_xlabel('Número de casos')
ax1.set_ylabel('Anos')
ax1.legend(loc='best')

#Furto de Veículos
df['furto_veiculos'].plot(kind='line', color='gray', linewidth=3, ax=ax2)
df['mms_furto_veiculos'].plot(kind='line',color='blue', linewidth=2, ax=ax2)

ax2.set_title('Furtos de Veículos')
ax2.set_xlabel('Número de casos')
ax2.set_ylabel('Anos')
ax2.legend(loc='best')

#Roubo em Coletivo
df['roubo_em_coletivo'].plot(kind='line', color='gray', linewidth=3, ax=ax3)
df['mms_roubo_em_coletivo'].plot(kind='line',color='blue', linewidth=2, ax=ax3)

ax3.set_title('Roubos em Coletivos')
ax3.set_xlabel('Número de casos')
ax3.set_ylabel('Anos')
ax3.legend(loc='best')

#Roubo Transeunte (Pessoas caminhando)
df['roubo_transeunte'].plot(kind='line', color='gray', linewidth=3, ax=ax4)
df['mms_roubo_transeunte'].plot(kind='line',color='blue', linewidth=2, ax=ax4)

ax4.set_title('Roubos a pessoas')
ax4.set_xlabel('Número de casos')
ax4.set_ylabel('Anos')
ax4.legend(loc='best')

plt.tight_layout()
```

A média Móvel Simples de cada coluna possui um comportamento próximo ao valor normal, tendo um comportamento mais suave. Demostra as tendências de aumento e queda.

Com relação ao comportamento do gráfico, a exceção do furto de veículos, os roubos sofretam um aumento significativo a partir de 2014.

Nos gráficos dos roubos, observam-se pequenas quedas em alguns momentos, porém é visível em todos eles uma queda um pouco mais acentuada em 2009, quando foi iniciado o programas das UPPs nas comunidades do Rio de Janeiro.

Cabe ressaltar que os dados consideram todo o estado do Rio de Janeiro, porém a região Metropolitada da Capital representa a maior parte dos dados.

###**Histograma**

Para finalizar a análise, serão plotados no mesmo gráfico os histogramas das ocorrências de homicídios culposo e doloso, permitindo uma comparação entre eles.

```
#Montando o Histogramas conjunto de Homicídios Doloso e Culposo
fig, ax = plt.subplots(figsize=(10,6))
ax.hist('hom_doloso', data=df, bins=15, color="blue")
ax.hist('hom_culposo', data=df, bins=15, color="red")
ax.set_title('Histograma dos Homicídios Culposos e Dolosos')
ax.set_ylabel('Frequência de Ocorrências (meses)')
ax.set_xlabel('Homicídios Culposos/Dolosos')
ax.legend(loc='best')
plt.tight_layout()
```
